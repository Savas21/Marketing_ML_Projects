```python
import pandas as pd
```

```python
df =  pd.read_csv("Data/ageinc.csv")
df.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(df["income"],df["age"])
plt.xlabel("Income")
plt.ylabel("Age")
plt.show()
```

```python
df["z_income"] = (df["income"]-df["income"].mean())/df["income"].std()
df["z_age"] = (df["age"]-df["age"].mean())/df["age"].std()
```

```python
from sklearn import cluster

colors = ["r","b","k","g","m","y"]
markers= ["^","o","d","s","p","X"]

plt.figure(figsize=(12,16))


for n in range(2,7):
    model = cluster.KMeans(n_clusters=n,random_state=10)
    df["cluster"] = model.fit_predict(df[["z_income","z_age"]])
    
    plt.subplot(3,2,n-1)
    for c in df["cluster"].unique():
        d =df[df["cluster"]==c]
        plt.scatter(d["income"],d["age"], marker=markers[c],color=colors[c])
        
plt.show()
```
